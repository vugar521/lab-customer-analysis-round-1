Read the three files into python as dataframes

```python
import pandas as pd
file1 = pd.read_csv('file1.csv')
file1.head()
```

```python
file2 = pd.read_csv('file2.csv')
file2.head()
```

```python
file3 = pd.read_csv('file3.csv')
file3.head()
```

Show the DataFrame's shape.

```python
print(file1.shape)
print(file2.shape)
print(file3.shape)
```

Standardize header names.

```python
file1.columns = file1.columns.str.replace(" ", "_")
file1.columns
```

```python
file1
```

```python
file2.columns = file2.columns.str.replace(" ", "_")
file2.columns
```

```python
file3.columns = file3.columns.str.replace(" ", "_")
file3.columns
```

Concatenate the three dataframes

```python
three_files = pd.concat([file1, file2, file3], axis=0) #discuss axis here
print(three_files.shape)
display(three_files)
```

Which columns are numerical?

```python
import numpy as np
three_files.select_dtypes(np.number)
```

```python
three_files.select_dtypes (['int32', 'float64'])
```

Which columns are categorical?  none of them !

```python
three_files.select_dtypes (['category'])
```

Perform the data cleaning operations mentioned so far in class

Delete the column education and the number of open complaints from the dataframe.

```python
three_files.columns
```

```python
three_files = three_files.drop(['Education','Total_Claim_Amount'], axis=1)
three_files.head()
```

Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
three_files['Customer_Lifetime_Value'] = three_files['Customer_Lifetime_Value'] *100
three_files.head ()
```

```python
three_files['Customer_Lifetime_Value'] =  pd.to_numeric(three_files['Customer_Lifetime_Value'], errors='coerce')
three_files.head ()
```

```python
three_files.drop_duplicates() #deleting duplicates
three_files
```
